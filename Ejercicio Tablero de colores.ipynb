### **Solución analítica**
La variable de estado es la posición de la ficha (verde, rojo, naranja,
azul)

Y los posibles estados en las monedas con las probabilidades:
*   cara - cara: 1/4
*   cara - cruz: 1/2
*   cruz - cara: 1/2
*   cruz - cruz: 1/4

Luego, denotamos:
*   **$x_n$** a la probabilidad de estar en el sector **naranja** tras **n** lanzamientos
*   **$y_n$** a la probabilidad de estar en el sector **azul** tras **n** lanzamientos
*   **$z_n$** a la probabilidad de estar en el sector **verde** tras **n** lanzamientos
*   **$t_n$** a la probabilidad de estar en el sector **rojo** tras **n** lanzamientos

Usando el teorema de la probabilidad total, en general nos queda:
$$
\begin{bmatrix}
x_n \\
y_n \\
z_n \\
t_n
\end{bmatrix}
=
\begin{bmatrix}
1/4 & 0 & 1/4 & 2/4 \\
2/4 & 1/4 & 0 & 1/4 \\
1/4 & 2/4 & 1/4 & 0 \\
0 & 1/4 & 2/4 & 1/4
\end{bmatrix}
\begin{bmatrix}
x_{n-1} \\
y_{n-1} \\
z_{n-1} \\
t_{n-1}
\end{bmatrix}
$$


### **Resolviendo por simulación:**
Sea  
- **rojo** la posición **0**
- **naranja** posición **1**
- **verde** posición **2**
- **azul** posición **3**


```python
import random
```

```python
# Empezamos en el color rojo asignado como posición 0
posicion = 0
conteo = [0, 0, 0, 0]  # contar cuántas veces cae en cada color
pasos = 1000  # pasos simulados
```

```python
for i in range(pasos):
    # Simulamos las 2 monedas donde 0 es cruz y 1 es cara
    # Usamos random.choice para que elija entre 0 y 1 (cruz o cara)
    moneda1 = random.choice([0, 1])
    moneda2 = random.choice([0, 1])
    avance = moneda1 + moneda2

    # Avanzar en el círculo
    posicion = (posicion + avance) % 4

    # Registrar visita
    conteo[posicion] += 1
```

```python
# Normalizar resultados para obtener frecuencias
frecuencias = [c / pasos for c in conteo]
```

```python
# Mostramos los resultados
print("Frecuencias por color [Rojo, Naranja, Verde, Azul]:")
print(frecuencias)
```

Gracias a nuestra simulación podemos concluir que las probabilidades de cada sector convergen a 0.25, lo que nos indica que, después de mucho tiempo y varios intentos, todos los sectores **si** son igual de probables.
