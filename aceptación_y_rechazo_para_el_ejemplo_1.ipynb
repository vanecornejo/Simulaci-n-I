```python
# Método de aceptación y rechazo para el ejemplo 1 visto en clase
# Importamos las librerías necesarias
import random as rd
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x)=2x
def f(x):
  return 2*x

# Definimos los límites del intervalo [a, b] y el valor máximo M de la función f(x)
a = 0
b = 1
M = 2

# Definimos la función de aceptación y rechazo
def aceptacion(n):
  lista = [] # lista para almacenar los datos "aceptados"
  for i in range(n):
    R1 = rd.random()
    R2 = rd.random()
    x = a + (b-a)*R1

# Utilizamos la condición if para aceptar o rechazar los valores
    if R2 <= f(x)/M: #criterio de aceptación
      lista.append(x) #guardamos los valores aceptados en la lista creada anteriormente
  return lista

# Mandamos llamar la función de aceptación y rechazo, con 100000 intentos
c = aceptacion(100000)
print(c) # imprimimos la lista con los valores aceptados

# Imprimimos el histograma con los valores aceptados
plt.hist(c, bins=100)

# Calculamos e imprimimos la media y la varianza de la muestra
print("La media es:", np.mean(c))
print("La varianza es:", np.var(c))

# Cuántos valores fueron aceptados
len(c)
```
