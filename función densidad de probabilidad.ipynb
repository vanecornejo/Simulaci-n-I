```python
from tabulate import tabulate
import matplotlib.pyplot as plt
import numpy as np

x0 = 5
a = 81
c = 89
m = 10**2

def generador(a,c,m,x0):
    x = (a*x0 + c) % m
    return x

#muestra = []
datos = []

for i in range(100):
    nuevo = generador(a,c,m,x0)
    datos.append(nuevo/m)
    #muestra.append([i,x0,nuevo/m])
    x0 = nuevo

#encabezados = ["i","x_i","x_{i+1}"]
#print(tabulate(muestra,headers=encabezados))

plt.plot(datos)
plt.hist(datos)
plt.show()

# verificando que se cumple el ciclo de vida, obtenemos la media y la varianza
media = np.mean(datos)
varianza = np.var(datos)
print("Media: ", media) # debería ser 1/2
print("Varianza: ", varianza) # debería ser 1/12

```
